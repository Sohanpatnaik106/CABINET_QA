# 5 Clusters

```python
import json
```

```python
from datasets import load_dataset
```

```python
with open("/home/sohanp/sccl/results/SCCL.distilroberta.DEBERTA_V3_CLASS.explicit.wikitq_train.question.lr1e-05.lrscale100.contrastive.eta1.tmp0.5.alpha1.0.seed0.clusters5/wikitq_cluster_assignment_1000.json", "r") as f:
    cluster_dict = json.load(f)
```

```python
train_dataset = load_dataset("wikitablequestions")["train"]
```

```python
cluster_labels_separation_dict = {}
```

```python
from tqdm import tqdm
```

```python
cluster_labels = {}
```

```python
for i, (key, value) in tqdm(enumerate(cluster_dict.items()), position = 0, leave = True, total = len(list(cluster_dict.keys()))):
    if value["cluster_label"] not in list(cluster_labels.keys()):
        cluster_labels[value["cluster_label"]] = {"id": [],
                                                  "question": [],
                                                  "table": [],
                                                  "answers": [],
                                                  "cluster_label": []}

    cluster_labels[value["cluster_label"]]["id"].append(i)
    cluster_labels[value["cluster_label"]]["question"].append(value["question"])
    cluster_labels[value["cluster_label"]]["table"].append(train_dataset[i]["table"])
    cluster_labels[value["cluster_label"]]["answers"].append(train_dataset[i]["answers"])
    cluster_labels[value["cluster_label"]]["cluster_label"].append(value["cluster_label"])
```

```python
import pandas as pd
for key, val in cluster_labels.items():
    df = pd.DataFrame(val)
    sampled_df = df.sample(frac = 0.025)
    sampled_df.reset_index(drop = True)

    sampled_df.to_csv(f"/home/sohanp/sccl/datasets/wikitq_train_cluster_{key}_sampled_0.25.csv", index = False)
```
